```python
#repeated sentences
from collections import Counter
import re

def count_repeated_sentences(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    
    # Split text into sentences using period, question mark, and exclamation mark as delimiters
    sentences = re.split(r'[.!?]', text)
    sentences = [sentence.strip() for sentence in sentences if sentence.strip()]
    
    sentence_counts = Counter(sentences)
    
    repeated_sentences = {sentence: count for sentence, count in sentence_counts.items() if count > 1}
    
    return repeated_sentences

file_path = "C:/Users/CHARAN/KGiSL/kgisl all glassdoor reviews.txt"
repeated_sentences = count_repeated_sentences(file_path)

# Sort the repeated sentences by count in descending order
sorted_repeated_sentences = dict(sorted(repeated_sentences.items(), key=lambda item: item[1], reverse=True))

for sentence, count in sorted_repeated_sentences.items():
    print(f"Sentence: '{sentence}' repeated {count} times.")
```

```python
#----------------- Finding the average of the ratings given by the users----------------#

def calculate_average(numbers):
    if not numbers:
        return 0
    
    total = sum(numbers)
    average = total / len(numbers)
    return average

file_path = "C:/Users/CHARAN/KGiSL/kgisl ratings extracted.txt"  # Replace with the actual path to your text file

# Read numbers from the file
numbers = []
with open(file_path, "r") as file:
    for line in file:
        try:
            number = float(line.strip())
            numbers.append(number)
        except ValueError:
            pass

# Calculate and print the average
average = calculate_average(numbers)
print(f"Average of the numbers: {average:.2f}")
```
